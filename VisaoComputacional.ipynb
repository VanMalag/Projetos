 **Redimensionar a Imagem**
##  Exercicio 3

```python
def redimensionar_imagem(percentual_scala, img):
  largura = int(img.shape[1] * percentual_scala / 200)
  altura = int(img.shape[0] * percentual_scala / 200)
  dim = (largura, altura)  
  img_redimensionada = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
  return img_redimensionada
     

red_img = redimensionar_imagem(20,img)
cv2_imshow(red_img)
     
```

```python
import numpy as np
import cv2
import os
from google.colab.patches import cv2_imshow
from skimage import data
import matplotlib.pyplot as plt
%matplotlib inline
     
```

```python
def redimensionar_imagem(percentual_scala, img):
  largura = int(img.shape[1] * percentual_scala / 100)
  altura = int(img.shape[0] * percentual_scala / 100)
  dim = (largura, altura)  
  img_redimensionada = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
  return img_redimensionada
```

```python
#Obter a Imagem
img = data.camera()  
print(f'tipo = {type(img)}')
```

```python
cv2_imshow(img)
```

```python
percentual_scala=20
redim_imagem=redimensionar_imagem(percentual_scala, img)
cv2_imshow(redim_imagem)
```
